```python
x_train = np.array([[2104,	5,	1,	45], [1416,	3,	2,	40	],[852,	2,	1,	35]])
y_train = np.array([460, 232, 178])
```

```python
def output_fwb(x,w,b):
  f_wb = np.dot(w,x) + b
  return f_wb
```

```python
print(output_fwb(x_train[0], [ 0.2 ,0., -0.01, -0.07], 0.00))
```

```python
def compute_cost(x,y,w,b):
  m = x.shape[0]
  cost = 0
  for i in range(m):
    temp_to_add = (output_fwb(x[i],w,b) - y[i] )**2
    cost = cost+ temp_to_add
  return cost/(2*m)
```

```python
def parameter_calculator(x,y, w,b):
  m,n = x.shape
  dj_dw = np.zeros(n)

  dj_db = 0

  for j in range(n):
    for i in range(m):
      dj_dw[j] = dj_dw[j] + (np.dot(x[i],w) + b - y[i])*(x[i][j])
      dj_db = dj_db + (np.dot(x[i],w) + b - y[i])
  return dj_dw/m, dj_db/m




```

```python
def gradient_descent(x,y,alpha):
  m,n = x.shape
  w = np.zeros(n)
  b = 0

  for i in range(1000):
    dj_dw, dj_db = parameter_calculator(x,y, w,b)
    temp_w = w
    temp_b = b
    temp_w = temp_w - alpha*dj_dw
    temp_b = temp_b - alpha*dj_db
    w = temp_w
    b = temp_b
  return w, b
```

```python
w, b = gradient_descent(x_train, y_train, 5.0e-7)
print(f"predicted_value = {np.dot(w,x_train[2]) + b}, actual_value = {y_train[2]}")
```
